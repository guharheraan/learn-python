```python
import numpy as np

students = {
    "Ali": [78, 85, 90, 66, 72],
    "Sara": [88, 92, 79, 85, 91],
    "Omar": [56, 60, 72, 65, 70]
}

# Function to calculate grade
def calculate_grade(avg):
    if avg >= 85:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 60:
        return "C"
    elif avg >= 50:
        return "D"
    else:
        return "F"

print("---- Individual Student Performance ----")
for name, marks in students.items():
    total = sum(marks)
    avg = total / len(marks)
    grade = calculate_grade(avg)

    print(f"\n{name}:")
    print(f"Marks: {marks}")
    print(f"Total Marks: {total}")
    print(f"Average Marks: {avg:.2f}")
    print(f"Grade: {grade}")

```

```python
# Class-Level Analysis (NumPy)

print("--- Class-Level Analysis ---")

marks_array = np.array(list(students.values()))
# print(marks_array)


subject_avg = np.mean(marks_array, axis=0)
subject_max = np.max(marks_array, axis=0)
subject_min = np.min(marks_array, axis=0)
subject_std = np.std(marks_array, axis=0)

# results
for i in range(marks_array.shape[1]):
    print(f"\nSubject {i+1}:")
    print(f"  Average = {subject_avg[i]:.2f}")
    print(f"  Highest = {subject_max[i]}")
    print(f"  Lowest  = {subject_min[i]}")
    print(f"  Std Dev = {subject_std[i]:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the images
img_white = np.ones((50, 50), dtype=np.uint8) * 255
img_black = np.zeros((50, 50), dtype=np.uint8)

# Merge images side by side
side_by_side = np.hstack((img_white, img_black))

# Merge images one above the other
one_above_other = np.vstack((img_white, img_black))

# Create a 2x2 checkerboard
top_row = np.hstack((img_white, img_black))
bottom_row = np.hstack((img_black, img_white))
checkerboard = np.vstack((top_row, bottom_row))

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(side_by_side, cmap='Dark2')
plt.title("Side by Side")

plt.subplot(1, 3, 2)
plt.imshow(one_above_other, cmap='Reds')
plt.title("One Above Other")

plt.subplot(1, 3, 3)
plt.imshow(checkerboard, cmap='Greens')
plt.title("Checkerboard")

plt.tight_layout()
plt.show()
```
